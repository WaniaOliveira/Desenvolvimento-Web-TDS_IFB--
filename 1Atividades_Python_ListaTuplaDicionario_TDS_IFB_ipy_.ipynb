````python
lista_de_tarefas = []
lista_de_tarefas.append({
    'id': 1,
    'descricao': 'Ir para Academia',
    'prioridade': 'Média',
    'concluida': False
})

lista_de_tarefas.append({
    'id': 2,
    'descricao': 'Estudar',
    'prioridade': 'Alta',
    'concluida': False
})

lista_de_tarefas.append({
    'id': 3,
    'descricao': 'Fazer almoço',
    'prioridade': 'Baixa',
    'concluida': False
})
print(f"Total de tarefas criadas: {len(lista_de_tarefas)}")
def exibir_tarefas(titulo, lista):
    print(f"\n--- {titulo}:({len(lista)}) tarefas)---")
    for t in lista:
        status = "concluida" if t ['concluida'] else "PENDENTE"
        print(f"ID: {t['id']} | Prioridade: {t['prioridade']:5s} | Status: {status} | Descrição: {t['descricao']}")
exibir_tarefas("Tarefas Atuais", lista_de_tarefas)
#Filtrando dados (Revisão do Exemplo 3: Filtragem)
#Utilizamos list comprehension (ou um loop if/else) para criar uma nova lista.
tarefas_pendentes = [
    tarefa for tarefa in lista_de_tarefas if tarefa['concluida'] == False
]

exibir_tarefas("Tarefas Pendentes", tarefas_pendentes)
id_para_concluir = 1
print(f"\n--- 3. Atualizando a Tarefa ID {id_para_concluir} ---")

# Iteramos na lista para encontrar o item correto.
for tarefa in lista_de_tarefas:
    # Busca pelo ID (Revisão do Exemplo 2: Busca por Rota)
    if tarefa['id'] == id_para_concluir:
        tarefa['concluida'] = True # O Update ocorre aqui!
        print(f"Status da tarefa '{tarefa['descricao']}' alterado para CONCLUÍDA.")
        break # É importante sair do loop assim que a tarefa for encontrada.

exibir_tarefas("Tarefas Após a Conclusão", lista_de_tarefas)
id_para_remover = 5
print(f"\n--- 4. Removendo a Tarefa ID {id_para_remover} ---")

# Encontrar o índice (posição) da tarefa na lista para usar o método .pop()
indice_para_remover = 5
for i, tarefa in enumerate(lista_de_tarefas):
    if tarefa['id'] == id_para_remover:
        indice_para_remover = i
        break

if indice_para_remover != 5:
    # Remove o item pelo seu índice (posição) na lista.
    tarefa_removida = lista_de_tarefas.pop(indice_para_remover)
    print(f"Tarefa removida: '{tarefa_removida['descricao']}'")
else:
    print(f"Erro: Tarefa com ID {id_para_remover} não encontrada.")

exibir_tarefas("Tarefas Após a Remoção", lista_de_tarefas)

# Mapeamento para garantir que 'Alta' venha antes de 'Baixa' na ordenação.
# Python ordena strings alfabeticamente, então criamos um valor numérico para forçar a ordem.
prioridade_ordem = {'Alta': 1, 'Média': 2, 'Baixa': 3}

# Usamos sorted() com lambda para ordenar a lista.
# A função key acessa o dicionário 'prioridade_ordem' para pegar o número de ordenação.
tarefas_ordenadas = sorted(
    lista_de_tarefas,
    key=lambda t: prioridade_ordem[t['prioridade']]
)

exibir_tarefas("Lista Final Ordenada por Prioridade", tarefas_ordenadas)
````

````python
#Exercicio 03

bandas = {
    "Gangrena Gasosa": [
        {"álbum": "Gente Ruim Só Manda Lembrança Pra Quem Não Presta", "ano": 2019},
        {"álbum": "Se Deus É 10, Satanás É 666", "ano": 1999},
        {"álbum": "Welcome to the Macumba", "ano": 1993}
    ],
    "Rogério Skylab": [
        {"álbum": "Nas Portas do Cu", "ano": 2023},
        {"álbum": "Crítica da Faculdade do Cu", "ano": 2021},
        {"álbum": "Cosmos", "ano": 2019}
    ],
    "Garotos Podres": [
        {"álbum": "Canções para Ninar", "ano": 2003},
        {"álbum": "Com a Corda Toda", "ano": 1997},
        {"álbum": "Rock de Subúrbio", "ano": 1995}
    ],
    "Massacration": [
        {"álbum": "Live Metal Espancation", "ano": 2017},
        {"álbum": "Good Blood Headbanguers", "ano": 2009},
        {"álbum": "Gates of Metal Fried Chicken of Death", "ano": 2005}
    ],
    "Raimundos": [
        {"álbum": "Cantigas de Roda", "ano": 2014},
        {"álbum": "Roda Viva", "ano": 2006},
        {"álbum": "Éramos 4", "ano": 2001}
    ]
}

# 1️- Encontrar o álbum mais antigo
mais_antigo = None #None é nada, é a falta de valor. No caso está imprimindo isso porque o resultado da função print() sempre é None porque ela é uma função que só faz uma ação e não gera um resultado válido
for banda, albuns in bandas.items():
    for album in albuns:
        if mais_antigo is None or album["ano"] < mais_antigo["ano"]:
            mais_antigo = {"banda": banda, "álbum": album["álbum"], "ano": album["ano"]}

print(f" Álbum mais antigo: '{mais_antigo['álbum']}' ({mais_antigo['ano']}) - {mais_antigo['banda']}")

# 2️- Banda com maior intervalo entre álbuns
maior_intervalo = 0
banda_maior_intervalo = ""

for banda, albuns in bandas.items():
    anos = [album["ano"] for album in albuns]
    intervalo = max(anos) - min(anos)
    if intervalo > maior_intervalo:
        maior_intervalo = intervalo
        banda_maior_intervalo = banda

print(f" Banda com maior intervalo entre álbuns: {banda_maior_intervalo} ({maior_intervalo} anos)")

# 3️- Todos os álbuns do mais novo para o mais antigo, com o nome da banda
todos_albuns = []

for banda, albuns in bandas.items():
    for album in albuns:
        todos_albuns.append({"banda": banda, "álbum": album["álbum"], "ano": album["ano"]})

ordenados = sorted(todos_albuns, key=lambda x: x["ano"], reverse=True)

print("\n--- Todos os álbuns do mais novo ao mais antigo ---")
for item in ordenados:
    print(f"{item['ano']} - {item['banda']}: {item['álbum']}")

````
