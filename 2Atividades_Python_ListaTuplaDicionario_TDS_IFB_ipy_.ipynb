```python
#Solicitar 5 números inteiros, armazenar em uma lista e imprimir.
# TAREFA 1
numeros = []  # Lista vazia para armazenar os números

for i in range(5):  # Repete 5 vezes
    n = int(input(f"Digite o {i+1}º número: "))  # Lê um número inteiro
    numeros.append(n)  # Adiciona à lista

print("Lista final:", numeros)
```

```python
#Aplicar o módulo (valor absoluto) em todos os elementos de uma lista.
# TAREFA 2
numeros = [-5, 3, -8, 2, -1]
modulos = [abs(n) for n in numeros]  # Usa a função abs() em cada número
print("Lista original:", numeros)
print("Lista com módulo aplicado:", modulos)
```

```python
#3- Apagar todos os elementos negativos de uma lista.
# TAREFA 3
numeros = [-5, 3, -8, 2, -1, 7]
positivos = [n for n in numeros if n >= 0]
print("Lista sem negativos:", positivos)
```

```python
#4- Ler notas até digitar número negativo, armazenar e mostrar média.
# TAREFA 4
notas = []

while True:
    nota = float(input("Digite uma nota (ou número negativo para sair): "))
    if nota < 0:
        break
    notas.append(nota)

if notas:  # Verifica se a lista não está vazia
    media = sum(notas) / len(notas)
    print(f"Média das notas: {media:.2f}")
else:
    print("Nenhuma nota foi informada.")
```

```python
# 5- Ler números positivos até digitar negativo.
#Separar pares e ímpares e mostrar a porcentagem de cada tipo.
# TAREFA 5
pares = []
impares = []
total = 0

while True:
    num = int(input("Digite um número positivo (ou negativo para sair): "))
    if num < 0:
        break
    total += 1
    if num % 2 == 0:
        pares.append(num)
    else:
        impares.append(num)

if total > 0:
    porc_pares = (len(pares) / total) * 100
    porc_impares = (len(impares) / total) * 100
    print(f"Porcentagem de pares: {porc_pares:.2f}%")
    print(f"Porcentagem de ímpares: {porc_impares:.2f}%")
else:
    print("Nenhum número foi informado.")
```

```python
# 6- Ler números positivos até digitar negativo e armazenar sem repetição.
# TAREFA 6
numeros = []

while True:
    n = int(input("Digite um número positivo (ou negativo para sair): "))
    if n < 0:
        break
    if n not in numeros:
        numeros.append(n)

print("Números sem repetição:", numeros)
```

```python
# 7- Solicitar altura e peso de 5 pessoas e verificar se pelo menos duas têm mesma altura e mesmo peso
# TAREFA 7
pessoas = []

for i in range(5):
    altura = float(input(f"Altura da pessoa {i+1}: "))
    peso = float(input(f"Peso da pessoa {i+1}: "))
    pessoas.append((altura, peso))  # Armazena como tupla (altura, peso)

repetidos = False
for i in range(len(pessoas)):
    for j in range(i + 1, len(pessoas)):
        if pessoas[i] == pessoas[j]:
            repetidos = True

if repetidos:
    print("Existem pelo menos duas pessoas com mesma altura e peso.")
else:
    print("Não há pessoas com mesma altura e peso.")
```

```python
# 8- Ler nome e idade de várias pessoas até nome vazio, e mostrar a mais velha.
# TAREFA 8
pessoas = []

while True:
    nome = input("Digite o nome (ou ENTER para sair): ")
    if nome == "":
        break
    idade = int(input("Digite a idade: "))
    pessoas.append((nome, idade))

if pessoas:
    mais_velha = max(pessoas, key=lambda x: x[1])
    print(f"A pessoa mais velha é {mais_velha[0]}, com {mais_velha[1]} anos.")
else:
    print("Nenhum dado informado.")
```
